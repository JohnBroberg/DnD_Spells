```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the spells page
url = "http://dnd2024.wikidot.com/spell:all"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")

# Find all tables on the page
tables = soup.find_all("table")

# Create an empty DataFrame
df = pd.DataFrame()

# Iterate over each table
for table in tables:
    # Get column titles from the first row (assuming they're in <th> tags)
    spell_titles = table.find_all("th")
    spell_table_titles = [title.text.strip() for title in spell_titles]

    # Extract all rows (excluding header)
    column_data = table.find_all("tr")[1:]

    # Extract each row's data
    for row in column_data:
        row_data = row.find_all("td")
        indiv_row_data = [data.text.strip() for data in row_data]

        # Append to DataFrame
        df = pd.concat([df, pd.DataFrame([indiv_row_data], columns=spell_table_titles)], ignore_index=True)

# Show the final DataFrame
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the spells page
url = "http://dnd2024.wikidot.com/spell:all"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")

# Find all tables and their corresponding headers
tables = soup.find_all("table")
headers = soup.find_all("em")  # Spell level headers in <em> tag

# Create an empty DataFrame
df = pd.DataFrame()

# Iterate over tables and extract spell level from headers
for index, table in enumerate(tables):
    # Get spell level from the corresponding header (if available)
    spell_level = headers[index].text.strip() if index < len(headers) else "Unknown"

    # Get column titles from the first row (assuming they're in <th> tags)
    spell_titles = table.find_all("th")
    spell_table_titles = [title.text.strip() for title in spell_titles]

    # Ensure "Spell Level" column exists
    spell_table_titles.append("Spell Level")

    # Extract all rows (excluding header)
    column_data = table.find_all("tr")[1:]

    # Extract each row's data
    for row in column_data:
        row_data = row.find_all("td")
        indiv_row_data = [data.text.strip() for data in row_data]
        indiv_row_data.append(spell_level)  # Add the spell level column

        # Append to DataFrame
        df = pd.concat([df, pd.DataFrame([indiv_row_data], columns=spell_table_titles)], ignore_index=True)

# Show the final DataFrame
print(df)
```
